```python
import keras
from keras import layers, models,optimizers
from keras.utils import to_categorical
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy import signal
from skimage import filters
from sklearn.metrics import accuracy_score


x_test = np.load("cifar100_x_test.npy")
y_test = np.load("cifar100_y_test.npy")


model_naive = models.load_model('cifar100_basic_cnn.keras')
model_overfit = models.load_model('cifar100_overfit_cnn.keras')
model_best = models.load_model('cifar100_third_cnn.keras')
model_vgg16 = models.load_model('cifar100_vgg16_cnn.keras')

model_naive.summary()
model_overfit.summary()
model_best.summary()
model_vgg16.summary()

naive_predictions = model_naive.predict(x_test)
overfit_predictions = model_overfit.predict(x_test)
best_predictions = model_best.predict(x_test)
vgg16_predictions = model_vgg16.predict(x_test)

naive_predictions = np.argmax(naive_predictions, axis=1)
overfit_predictions = np.argmax(overfit_predictions, axis=1)
best_predictions = np.argmax(best_predictions, axis=1)
vgg16_predictions = np.argmax(vgg16_predictions, axis=1)



print("Naive test accuracy: ", str(accuracy_score(naive_predictions, y_test)))
print("Overfit test accuracy: ", str(accuracy_score(overfit_predictions, y_test)))
print("Third test accuracy: ", str(accuracy_score(best_predictions, y_test)))
print("Transfer VGG16 test accuracy: ", str(accuracy_score(vgg16_predictions, y_test)))


```
